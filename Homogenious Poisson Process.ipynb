```python
import numpy as np
import matplotlib.pyplot as plt


# Simulación proceso de poisson homogéneo

max_time = 100
nu = 0.5
contador = {
    'T': [0],
    'N': [0]
}
t = 0
i = 0 
while t <= max_time:
    u = np.random.random()
    t = t - (1/nu) * np.log(u)
    i += 1
    contador['T'].append(t)
    contador['N'].append(i)

plt.figure(figsize=(10, 6))
plt.scatter(x = contador['T'], y = contador['N'], edgecolor='black', alpha=0.7)
plt.step(contador['T'],contador['N']   )
plt.xlabel("Tiempo")
plt.ylabel('Eventos')
plt.title(f"Eventos en un proceso de Poisson homogeneo de tiempo {max_time} y lambda = {nu}")
plt.legend()
plt.show()
```
